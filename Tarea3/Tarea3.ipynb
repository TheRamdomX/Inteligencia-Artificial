```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import f1_score, confusion_matrix, classification_report

# Ruta donde están las carpetas por clase
DATA_DIR = "/Imagenes"

IMG_SIZE = (64, 64)
BATCH_SIZE = 32
SEED = 42

# Carga con split automático

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.30,     # primero separar train (70%) + lo que queda para val + test
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

temp_ds = tf.keras.preprocessing.image_dataset_from_directory(
    DATA_DIR,
    validation_split=0.30,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# Ahora separar validación (15%) y test (15%) del 30% restante

val_ds = temp_ds.take(len(temp_ds)//2)
test_ds = temp_ds.skip(len(temp_ds)//2)

class_names = train_ds.class_names
num_classes = len(class_names)
print("Clases detectadas:", class_names)

# Prefetch
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(AUTOTUNE)
val_ds   = val_ds.prefetch(AUTOTUNE)
test_ds  = test_ds.prefetch(AUTOTUNE)

# Normalización

normalization_layer = layers.Rescaling(1./255)

def normalize(ds):
    return ds.map(lambda x, y: (normalization_layer(x), y))

train_ds = normalize(train_ds)
val_ds   = normalize(val_ds)
test_ds  = normalize(test_ds)

# Arquitectura Base CNN

def build_base_model():
    model = keras.Sequential([
        layers.Conv2D(32, (3,3), padding="same", activation="relu", input_shape=IMG_SIZE + (3,)),
        layers.MaxPooling2D(),

        layers.Conv2D(64, (3,3), padding="same", activation="relu"),
        layers.MaxPooling2D(),

        layers.Conv2D(128, (3,3), padding="same", activation="relu"),
        layers.MaxPooling2D(),

        layers.Flatten(),
        layers.Dense(256, activation="relu"),
        layers.Dense(num_classes, activation="softmax")
    ])

    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    return model

model_base = build_base_model()
model_base.summary()

# Arquitectura con Dropout

def build_dropout_model():
    model = keras.Sequential([
        layers.Conv2D(32, (3,3), padding="same", activation="relu", input_shape=IMG_SIZE + (3,)),
        layers.MaxPooling2D(),

        layers.Conv2D(64, (3,3), padding="same", activation="relu"),
        layers.MaxPooling2D(),

        layers.Conv2D(128, (3,3), padding="same", activation="relu"),
        layers.MaxPooling2D(),

        layers.Dropout(0.25),   # Dropout tras convolucionales

        layers.Flatten(),
        layers.Dense(256, activation="relu"),
        layers.Dropout(0.5),    #  Dropout antes del FC final
        layers.Dense(num_classes, activation="softmax")
    ])

    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    return model

model_dropout = build_dropout_model()
model_dropout.summary()

# Entrenamiento

EPOCHS = 30

history_base = model_base.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
)

history_dropout = model_dropout.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS
)

# Curvas de Loss y Accuracy

def plot_history(history, title="Modelo"):
    plt.figure(figsize=(12,5))

    # Loss
    plt.subplot(1,2,1)
    plt.plot(history.history["loss"], label="train")
    plt.plot(history.history["val_loss"], label="val")
    plt.title(f"{title} - Loss")
    plt.xlabel("Época")
    plt.ylabel("Loss")
    plt.legend()

    # Accuracy
    plt.subplot(1,2,2)
    plt.plot(history.history["accuracy"], label="train")
    plt.plot(history.history["val_accuracy"], label="val")
    plt.title(f"{title} - Accuracy")
    plt.xlabel("Época")
    plt.ylabel("Accuracy")
    plt.legend()

    plt.show()

plot_history(history_base, "Base")
plot_history(history_dropout, "Dropout")

# Evaluación en Test (Accuracy + F1)

def evaluate_model(model, ds, name):
    y_true = []
    y_pred = []

    for images, labels in ds:
        preds = model.predict(images)
        preds = np.argmax(preds, axis=1)
        y_true.extend(labels.numpy())
        y_pred.extend(preds)

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    acc = np.mean(y_true == y_pred)
    f1 = f1_score(y_true, y_pred, average="macro")

    print(f"\n===== Resultados de {name} =====")
    print("Accuracy:", acc)
    print("F1-macro:", f1)
    print("\nMatriz de confusión:")
    print(confusion_matrix(y_true, y_pred))
    print("\nReporte por clase:")
    print(classification_report(y_true, y_pred, target_names=class_names))

    return acc, f1

acc_b, f1_b = evaluate_model(model_base, test_ds, "Modelo Base")
acc_d, f1_d = evaluate_model(model_dropout, test_ds, "Modelo Dropout")
```
