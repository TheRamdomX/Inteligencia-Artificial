# Estimación de π usando el Filtro Monte Carlo

El filtro Monte Carlo es un método probabilístico que utiliza números aleatorios para resolver problemas que pueden ser deterministas en principio. En este caso, se utiliza para estimar el valor de π. El método consiste en generar puntos aleatorios dentro de un cuadrado de lado 1 y contar cuántos caen dentro de un cuarto de círculo inscrito en ese cuadrado. La proporción de puntos dentro del círculo respecto al total, multiplicada por 4, aproxima el valor de π. A mayor cantidad de muestras, la estimación es más precisa.


```python
import numpy as np
import matplotlib.pyplot as plt

def estimate_pi(n_samples):
    x = np.random.rand(n_samples)
    y = np.random.rand(n_samples)
    inside_circle = (x**2 + y**2) <= 1
    pi_estimate = 4 * np.sum(inside_circle) / n_samples
    return x, y, inside_circle, pi_estimate

# Ejecutar dos procesos
n_samples1 = 10000
n_samples2 = n_samples1 // 5

x1, y1, inside1, pi_estimate1 = estimate_pi(n_samples1)
x2, y2, inside2, pi_estimate2 = estimate_pi(n_samples2)

# Gráficos en subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].scatter(x1[inside1], y1[inside1], s=1, c="blue", label="Dentro del círculo")
axs[0].scatter(x1[~inside1], y1[~inside1], s=1, c="red", label="Fuera del círculo")
circle1 = plt.Circle((0,0), 1, color="black", fill=False)
axs[0].add_patch(circle1)
axs[0].set_aspect("equal")
axs[0].set_title(f"n={n_samples1}\nπ ≈ {pi_estimate1:.4f}")
axs[0].legend()

axs[1].scatter(x2[inside2], y2[inside2], s=1, c="blue", label="Dentro del círculo")
axs[1].scatter(x2[~inside2], y2[~inside2], s=1, c="red", label="Fuera del círculo")
circle2 = plt.Circle((0,0), 1, color="black", fill=False)
axs[1].add_patch(circle2)
axs[1].set_aspect("equal")
axs[1].set_title(f"n={n_samples2}\nπ ≈ {pi_estimate2:.4f}")
axs[1].legend()

plt.suptitle("Estimación Monte Carlo de π con dos cantidades de muestras")
plt.tight_layout()
plt.show()
```

Al ejecutar el código, se muestran dos gráficos: cada uno representa la simulación con diferente cantidad de muestras. Los puntos azules están dentro del círculo y los rojos fuera. Se observa que, al aumentar el número de muestras, la estimación de π mejora y la distribución de puntos se ajusta mejor a la forma del círculo. Esto ilustra cómo el método Monte Carlo se vuelve más preciso con más datos.
