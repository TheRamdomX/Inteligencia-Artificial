```python
from sklearn import datasets
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt

# Cargar el dataset Iris
iris = datasets.load_iris()
X = iris.data[:, :2]  # Solo tomamos las dos primeras características
y = iris.target

# Crear y ajustar el modelo MeanShift
mean_shift = MeanShift()
mean_shift.fit(X)

# Obtener resultados
labels = mean_shift.labels_
cluster_centers = mean_shift.cluster_centers_

# Visualización
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=300, c='red', marker='X', label='Centros')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Clustering con MeanShift')
plt.legend()
plt.show()

print(f"Cantidad de clusters detectados: {len(cluster_centers)}")
```

```python
from sklearn.cluster import estimate_bandwidth

# Calcular un bandwidth más pequeño
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=len(X))

mean_shift = MeanShift(bandwidth=bandwidth)
mean_shift.fit(X)

labels = mean_shift.labels_
cluster_centers = mean_shift.cluster_centers_

print(f"Clusters detectados: {len(cluster_centers)}")

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(cluster_centers[:,0], cluster_centers[:,1], s=300, c='red', marker='X')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('MeanShift con bandwidth ajustado')
plt.show()
```
