```python
import matplotlib.pyplot as plt

from sklearn.cluster import kmeans_plusplus
from sklearn import datasets

iris = datasets.load_iris()

X = iris.data[:, :2]
y = iris.target

# Calculate seeds from k-means++ (returns a tuple: (centers, indices))
centers_init, indices = kmeans_plusplus(X, n_clusters=3, random_state=0)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(centers_init[:, 0], centers_init[:, 1], s=250, c='red', marker='.')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')  
plt.show()
```

```python
from sklearn.cluster import KMeans

# Crear y ajustar el modelo K-Means
kmeans = KMeans(n_clusters=3, init=centers_init, n_init=1, random_state=0)
kmeans.fit(X)

# Obtener los centros finales y las etiquetas asignadas
final_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualizar resultados
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(centers_init[:, 0], centers_init[:, 1], s=100, c='black', marker='o', label='Centros iniciales')
plt.scatter(final_centers[:, 0], final_centers[:, 1], s=100, c='red', marker='X', label='Centros finales')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Comparacin entre centros iniciales y finales (K-Means++)')
plt.legend()
plt.show()
```
